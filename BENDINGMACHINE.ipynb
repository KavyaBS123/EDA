## Identify and handle missing values


```python
missing_values = df.isnull().sum()
display(missing_values)
```

```python
for col in ['B_PowerAngle', 'Frequency', 'Total_ImportEnergyConsumption', 'Total_ExportEnergyConsumption']:
    df[col].fillna(df[col].mean(), inplace=True)

missing_values_after_imputation = df.isnull().sum()
display(missing_values_after_imputation)
```

## Identify and handle outliers




```python
numerical_cols = df.select_dtypes(include=['number']).columns

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap outliers
    df[col] = df[col].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))

# Verify outliers handling
outliers_after_handling = {}
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    outliers_after_handling[col] = outliers_count

display("Outliers count after capping:")
display(outliers_after_handling)
```
