PERFORMING FEATURE ENGINEERING

```python
df_moulding['TimeStamp']=pd.to_datetime(df_moulding['TimeStamp'])
```

EXTRACTING TIME-BASED FEATURES

```python
df_moulding['Hour']=df_moulding['TimeStamp'].dt.hour
df_moulding['DayOfWeek']=df_moulding['TimeStamp'].dt.dayofweek
df_moulding['Month']=df_moulding['TimeStamp'].dt.month
display(df_moulding.head())
```

CALCULATING THE RATE OF CHANGE FOR 'Total_ImportEnergyConsumption'  

```python
df_moulding['Energy_Consumption_Rate_of_Change'] = df_moulding['Total_ImportEnergyConsumption'].diff()

df_moulding['DigitalInput1_Pulses_Interaction'] = df_moulding['DigitalInput1'] * df_moulding['NumberPulses_DigitalInput1']
df_moulding['DigitalInput2_Pulses_Interaction'] = df_moulding['DigitalInput2'] * df_moulding['NumberPulses_DigitalInput2']

display(df_moulding.head())
```

PERFORMING CATEGORICAL FEATURE ENCODING

identifying if any there and converting it , dropping the DeviceId and TimeStamp

```python
categorical_cols = df_moulding.select_dtypes(include='object').columns
print("Categorical columns:", categorical_cols)
df_moulding = df_moulding.drop(columns=['DeviceId', 'TimeStamp'])

display(df_moulding.head())
```

FEATURE SCALING - to ensure they are in similar scale

```python
from sklearn.preprocessing import StandardScaler

numerical_cols = df_moulding.select_dtypes(include=['float64', 'int64']).columns.tolist()
cols_to_scale = ['AnalogInput2', 'VRN', 'VYN', 'VBN', 'Current_R', 'Current_Y', 'Current_B',
                 'R_PowerFactor', 'Y_PowerFactor', 'B_PowerFactor', 'R_PowerAngle', 'Y_PowerAngle',
                 'B_PowerAngle', 'Frequency', 'Total_ImportEnergyConsumption', 'Current']


scaler = StandardScaler()

df_moulding[cols_to_scale] = scaler.fit_transform(df_moulding[cols_to_scale])

display(df_moulding.head())
```
