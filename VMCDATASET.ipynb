## Load the dataset


```python
import pandas as pd
df = pd.read_excel('/content/VMC_machine.xlsx')
```

```python
df.head()
```

## Explore the data



```python
df.info()
df.describe()
```

## Check for missing values



```python
import matplotlib.pyplot as plt
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)

if not missing_values.empty:
    plt.figure(figsize=(10, 6))
    missing_values.plot(kind='bar')
    plt.title('Number of Missing Values per Column')
    plt.xlabel('Columns')
    plt.ylabel('Number of Missing Values')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("No missing values found.")
```

## Visualize the data




```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select key numerical columns, using available voltage columns
key_numerical_cols = ['Current', 'VRN', 'VYN', 'VBN', 'Frequency', 'Total_ImportEnergyConsumption']

# Plot histograms for selected columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(key_numerical_cols):
    plt.subplot(2, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Plot scatter plots for pairs of selected columns
plt.figure(figsize=(15, 10))
sns.pairplot(df[key_numerical_cols])
plt.suptitle('Pairwise Relationships of Key Numerical Columns', y=1.02)
plt.show()
```
