### Identify and Handle Outliers

```python

numerical_cols = ['Current', 'Frequency', 'Total_ImportEnergyConsumption', 'VRN', 'Current_R', 'Current_Y', 'Current_B', 'VRN', 'VYN', 'VBN', 'R_PowerFactor', 'Y_PowerFactor', 'B_PowerFactor', 'R_PowerAngle', 'Y_PowerAngle', 'B_PowerAngle', 'Total_ExportEnergyConsumption']

plt.figure(figsize=(15, 20))
for i, col in enumerate(numerical_cols):
    plt.subplot(6, 3, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
plt.tight_layout()
plt.show()

Q1 = df['Current'].quantile(0.25)
Q3 = df['Current'].quantile(0.75)
IQR = Q3 - Q1

df_cleaned = df[~((df['Current'] < (Q1 - 1.5 * IQR)) | (df['Current'] > (Q3 + 1.5 * IQR)))]

print(f"Original shape: {df.shape}")
print(f"Shape after outlier removal in 'Current': {df_cleaned.shape}")
```

TIME STAMP CONVERSION


```python
df_cleaned['TimeStamp'] = pd.to_datetime(df_cleaned['TimeStamp'])
display(df_cleaned.info())
```

## Extract time-based features



```python
df_cleaned['hour'] = df_cleaned['TimeStamp'].dt.hour
df_cleaned['day_of_week'] = df_cleaned['TimeStamp'].dt.dayofweek
df_cleaned['month'] = df_cleaned['TimeStamp'].dt.month
display(df_cleaned.head())
```

## Create lagged features
Create lagged features for relevant numerical columns to capture temporal dependencies.



Create lagged features for the numerical columns and handle the resulting missing values.



```python
numerical_cols_for_lag = ['Current', 'Frequency', 'VRN', 'Current_R', 'Total_ImportEnergyConsumption']

for col in numerical_cols_for_lag:
    df_cleaned[f'{col}_lag1'] = df_cleaned[col].shift(1)
    df_cleaned[f'{col}_lag2'] = df_cleaned[col].shift(2)


df_cleaned.dropna(inplace=True)

display(df_cleaned.head())
display(df_cleaned.isnull().sum())
```

## Create rolling window features


```python
numerical_cols_for_rolling = ['Current', 'Frequency', 'VRN', 'Current_R', 'Total_ImportEnergyConsumption']
window_size = 5

for col in numerical_cols_for_rolling:
    df_cleaned[f'{col}_rolling_mean'] = df_cleaned[col].rolling(window=window_size).mean()
    df_cleaned[f'{col}_rolling_std'] = df_cleaned[col].rolling(window=window_size).std()

df_cleaned.dropna(inplace=True)

display(df_cleaned.head())
display(df_cleaned.isnull().sum())
```

## Polynomial features


```python
from sklearn.preprocessing import PolynomialFeatures

numerical_cols_for_poly = ['Current', 'VRN', 'Frequency']

poly = PolynomialFeatures(degree=2, include_bias=False)

poly_features = poly.fit_transform(df_cleaned[numerical_cols_for_poly])

poly_feature_names = poly.get_feature_names_out(numerical_cols_for_poly)
poly_df = pd.DataFrame(poly_features, columns=poly_feature_names, index=df_cleaned.index)

df_cleaned = pd.concat([df_cleaned, poly_df], axis=1)

display(df_cleaned.head())
```

## Interaction features


```python
df_cleaned['Current_R_VRN_interaction'] = df_cleaned['Current_R'] * df_cleaned['VRN']
df_cleaned['Current_TotalEnergy_interaction'] = df_cleaned['Current'] * df_cleaned['Total_ImportEnergyConsumption']

display(df_cleaned.head())
```

```python
df_cleaned = df_cleaned.reset_index(drop=True)

df_cleaned['Current_R_VRN_interaction'] = df_cleaned['Current_R'] * df_cleaned['VRN']
df_cleaned['Current_TotalEnergy_interaction'] = df_cleaned['Current'] * df_cleaned['Total_ImportEnergyConsumption']

display(df_cleaned.head())
```

```python
print(df_cleaned.columns[df_cleaned.columns.duplicated()])

df_cleaned = df_cleaned.loc[:, ~df_cleaned.columns.duplicated()]

df_cleaned['Current_R_VRN_interaction'] = df_cleaned['Current_R'] * df_cleaned['VRN']
df_cleaned['Current_TotalEnergy_interaction'] = df_cleaned['Current'] * df_cleaned['Total_ImportEnergyConsumption']

display(df_cleaned.head())
```

## Feature scaling


```python
from sklearn.preprocessing import StandardScaler

numerical_cols_to_scale = df_cleaned.select_dtypes(include=['float64', 'int64']).columns.tolist()
numerical_cols_to_scale = [col for col in numerical_cols_to_scale if col not in ['DigitalInput1', 'DigitalInput2', 'NumberPulses_DigitalInput1', 'NumberPulses_DigitalInput2', 'hour', 'day_of_week', 'month']]

scaler = StandardScaler()

df_cleaned[numerical_cols_to_scale] = scaler.fit_transform(df_cleaned[numerical_cols_to_scale])

display(df_cleaned.head())
display(df_cleaned[numerical_cols_to_scale].describe())
```
